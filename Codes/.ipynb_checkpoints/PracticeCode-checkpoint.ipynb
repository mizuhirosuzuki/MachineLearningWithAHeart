# Build the Model

## Logistic Regression

```python
# for preprocessing the data
from sklearn.preprocessing import StandardScaler

# the model
from sklearn.linear_model import LogisticRegression

# for combining the preprocess with model training
from sklearn.pipeline import Pipeline

# for optimizing parameters of the pipeline
from sklearn.model_selection import GridSearchCV
```

In Pipeline, you pass a list of "steps"in the form of tuples with a step name in the first field and the object associated to the step in the second.
Pipeline then creates one "estimator" object that you can pass data into for training and prediction.

```python
pipe = Pipeline(steps = [('scale', StandardScaler()),
                        ('logistic', LogisticRegression())])
pipe
```

Grid search allows you try out different parameters in the pipeline.
Below we try out different values for Scikit Learn's logistic regression "C" parameters as well as its regularization method.

The CV in GridSearchCV is anotehr best practice to prevent overfitting your model.

```python
param_grid = {'logistic__C' : np.linspace(0.0001, 10, num = 100),
             'logistic__penalty' : ['l1', 'l2']}
gs = GridSearchCV(estimator = pipe,
                 param_grid = param_grid,
                 cv = 3)
```

```python
gs
```

With the parameter grid we've created and cross-valication, we're about to test 30 different models and take the best one.

```python
gs.fit(train_values_subset, train_labels.heart_disease_present)
```

Let's take a look at the best parametesr.

```python
gs.best_params_
```

And the in-sample log loss score.
Notice that since log loss wants the class probabilities we call predict_proba and not simply predict, which would return the predicted labels, not their probabilities.

```python
from sklearn.metrics import log_loss

in_sample_preds = gs.predict_proba(train_values[selected_features])
log_loss(train_labels.heart_disease_present, in_sample_preds)
```
