# Prediction accuracies of different methods

## Preparation

First, I clean datasets, which are used for all the analyses below.

```python
# Import libraries
%matplotlib inline

import numpy as np
import pandas as pd
import itertools
import random

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
```

I load the datasets:

```python
# Load datasets for training my model
train_values = pd.read_csv('../Data/train_values.csv', index_col = 'patient_id')
train_labels = pd.read_csv('../Data/train_labels.csv', index_col = 'patient_id')
```

The head of the training features looks like this:

```python
train_values.head()
```

```python
# convert a categorical variable 'thal' into one-hot coding
train_values = pd.get_dummies(train_values, columns = ['thal'])
```

```python
train_values.head()
```

```python
# split training and test data (75:25)
# test data is used to evaluate each method
train_values, test_values, train_labels, test_labels = train_test_split(train_values, 
                                                                        train_labels.heart_disease_present,
                                                                       test_size = 0.25)
```

## Logistic regression

```python
# for preprocessing the data
from sklearn.preprocessing import StandardScaler

# the model
from sklearn.linear_model import LogisticRegression

# for combining the preprocess with model training
from sklearn.pipeline import Pipeline

# for optimizing parameters of the pipeline
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
pipeline_logit = Pipeline(steps = [('scale', StandardScaler()),
                                   ('logistic', LogisticRegression(solver = 'liblinear'))])
```

```python
# grid search for hyper parameters
param_grid = {'logistic__C': np.linspace(0.0001, 10, num = 100),
              'logistic__penalty': ['l1', 'l2']}
gs_logit = GridSearchCV(estimator = pipeline_logit,
                        param_grid = param_grid,
                        cv = 5,
                        scoring='neg_log_loss')
```

```python
gs_logit.fit(train_values, train_labels)
```

```python
# model evaluation

pred_logit = gs_logit.predict_proba(test_values)
score_logit = log_loss(test_labels, pred_logit)
print("Log loss score: %.3f" % score_logit)
```

### Lasso

```python
n_alphas = 300
alphas = np.logspace(-3, 1, n_alphas)

coefs = []
for a in alphas:
    lasso = LogisticRegression(C = a, penalty = 'l1', solver = 'liblinear')
    lasso.fit(train_values, train_labels)
    coefs.append(lasso.coef_.tolist()[0])

ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('C')
plt.ylabel('weights')
plt.title('Lasso coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

### Ridge

```python
n_alphas = 300
alphas = np.logspace(-3, 1, n_alphas)

coefs = []
for a in alphas:
    ridge = LogisticRegression(C = a, penalty = 'l2', solver = 'liblinear')
    ridge.fit(train_values, train_labels)
    coefs.append(ridge.coef_.tolist()[0])

ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('C')
plt.ylabel('weights')
plt.title('ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

### Elastic net

```python
pipeline_en = Pipeline(steps = [('scale', StandardScaler()),
                               ('logistic', LogisticRegression(penalty = 'elasticnet', solver = 'saga'))])
```

```python
# grid search for hyper parameters
param_grid = {'logistic__C': np.linspace(0.0001, 10, num = 100),
              'logistic__l1_ratio': np.linspace(0.0001, 0.9999, num = 10)}
gs_en = GridSearchCV(estimator = pipeline_en,
                        param_grid = param_grid,
                        cv = 5,
                        scoring='neg_log_loss')
```

```python
gs_en.fit(train_values, train_labels)
```

```python
# model evaluation

pred_en = gs_en.predict_proba(test_values)
score_en = log_loss(test_labels, pred_en)
print("Log loss score: %.3f" % score_en)
```

```python
# model evaluation

pred_tree = gs_tree.predict_proba(test_values)
score_tree = log_loss(test_labels, pred_tree)
print("Log loss score: %.3f" % score_tree)
```

## Deep neural net

```python
import tensorflow as tf
from keras import datasets, layers, models
from keras.callbacks import EarlyStopping, ModelCheckpoint, LambdaCallback, Callback
from keras.models import Sequential, Model, load_model
from keras.layers import Flatten, Input, concatenate
from keras.layers.core import Activation, Dropout, Dense
from keras.optimizers import Adam
from keras.utils import plot_model

# for combining the preprocess with model training
from sklearn.pipeline import Pipeline

# for optimizing parameters of the pipeline
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
total_dim_var = train_values.shape[1]
```

```python
model_mlp = Sequential()
model_mlp.add(Dense(16, input_dim = total_dim_var, activation = 'relu'))
for i in range(2):
  model_mlp.add(Dense(16, input_dim = 16, activation = 'relu'))
model_mlp.add(Dense(1, input_dim = 16, activation = 'sigmoid'))

model_mlp.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
plot_model(model_mlp, show_shapes = True)
```

```python
X_st = StandardScaler()
train_values_st = X_st.fit_transform(train_values)
test_values_st = X_st.transform(test_values)
```

```python
model_mlp.fit(train_values_st, 
              train_labels, 
              epochs = 30)
```

```python
# model evaluation

pred_dnn = model_mlp.predict_proba(test_values_st)
score_dnn = log_loss(test_labels, pred_dnn)
print("Log loss score: %.3f" % score_dnn)
```
